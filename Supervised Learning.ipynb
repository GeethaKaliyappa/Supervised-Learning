#### 3E) Balance the data using the right balancing technique. [2 Marks]

    i.   Check distribution of the target variable
    ii.  Say output is class A : 20% and class B : 80%
    iii. Here you need to balance the target variable as 50:50.
    iv.  Try appropriate method to achieve the same.

```python
D7['LoanOnCard'].value_counts()
```

```python
sns.countplot(D7['LoanOnCard']);
```

Highly imbalanced data. Need to balance it

```python
D8 = X_train

D8['LoanOnCard'] = y_train

D8.head()
```

```python
D8['LoanOnCard'].value_counts()
```

```python
#Downsample the majority Class 

df_majority = D8[(D8['LoanOnCard']==False)] 
df_minority = D8[(D8['LoanOnCard']==True)]

#Checking the shape post split

print('Major Class Shape: ', df_majority.shape)
print('Minor Class Shape: ', df_minority.shape)

from sklearn.utils import resample

df_resampled = resample(df_majority, 
                        replace=True,      
                        n_samples= 366,   
                        random_state=42)   
```

```python
D9 = pd.concat([df_resampled, df_minority])

print('Shape after resampling: ', D9.shape)
```

```python
sns.countplot(D9['LoanOnCard']);
```

```python
X_train = D9.drop(labels= "LoanOnCard" , axis = 1)
y_train = D9["LoanOnCard"]
```

#### 4A) Train a base model each for SVM, KNN. [4 Marks]

```python
from sklearn import svm

# Building SVM Model

SVM = svm.SVC()  

SVM.fit(X_train, y_train)

#Test Model

y_predict = SVM.predict(X_test)

#check-Accuracy

SVM.score(X_test, y_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier

# Building KNN Model

NNH = KNeighborsClassifier()

# Fit KNN Model

NNH.fit(X_train, y_train)

#Test the Model

y_predict = NNH.predict(X_test)

#check score

NNH.score(X_test, y_test)
```

#### 4B) Tune parameters for each of the models wherever required and finalize a model. [3 Marks]

(Optional: Experiment with various Hyperparameters - Research required)

```python
from sklearn import svm

# Tuning SVM Model on Upsampeld Data

SVM = svm.SVC(gamma=0.02, C=1)  

# Fit SVM Model

SVM.fit(X_train, y_train)

#Test the Performance of the Model

y_predict = SVM.predict(X_test)

SVM.score(X_test, y_test)
```

```python
print(metrics.classification_report(y_test, y_predict))
```

```python
from sklearn.neighbors import KNeighborsClassifier

# Tuning KNN Model on Upsampeld Data

NNH = KNeighborsClassifier(n_neighbors=3, p=2)

# Fit KNN Model

NNH.fit(X_train, y_train)

#Test the Performance of the Model

y_predict = NNH.predict(X_test)

NNH.score(X_test, y_test)
```

```python
print(metrics.classification_report(y_test, y_predict))
```

```python
from sklearn.linear_model import LogisticRegression

# Tuning Log-Regression Model on Upsampeld Data

LR = LogisticRegression(penalty='none', solver='lbfgs')
LR.fit(X_train, y_train)

#predict on test
y_predict = LR.predict(X_test)


LR.score(X_test, y_test)
```

```python
print(metrics.classification_report(y_test, y_predict))
```

Accuracy for Logistic Regression is comparitively higher than SVM and KNN. Also the precision and recall are also good. Hence going ahead with LR Model.
